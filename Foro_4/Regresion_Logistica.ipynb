```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
```

```python
df = pd.read_csv("C:/Users/Unicom-7344/Desktop/Emiliano_Palacios/Uni 2/Aprendizaje Automatico/usuarios_win_mac_lin.csv")
df.head()
```

```python
df.info()
df.describe()
```

```python
plt.scatter(df['duracion'], df['paginas'], c=df['clase'])
plt.xlabel("duración")
plt.ylabel("páginas")
plt.title("Distribución de usuarios según sistema operativo")
plt.show()
```

```python
X = df[['duracion', 'paginas', 'acciones', 'valor']]
Y = df['clase']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=0)
```

```python
from sklearn.linear_model import LogisticRegression

modelo = LogisticRegression(max_iter=1000)
modelo.fit(X_train, Y_train)
```

```python
Y_pred = modelo.predict(X_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix

predictions = modelo.predict(X_test)

print("Precisión: ", modelo.score(X_test, Y_test))

print(classification_report(Y_test, predictions))

print(confusion_matrix(Y_test, predictions))
```
