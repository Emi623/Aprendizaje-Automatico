## Ejercicio 6
**Realizar la suma elemento a elemento de los arrays**  
`a = [1,2,3]` y `b = [4,5,6]`.

```python
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])

suma = a + b

print("Ejercicio 6:")
print("Array a:", a)
print("Array b:", b)
print("Suma elemento a elemento:", suma)
```

## Ejercicio 7
**Multiplicar dos matrices 3x3 de enteros aleatorios entre 1 y 10.**

```python
matriz1 = np.random.randint(1, 11, (3, 3))
matriz2 = np.random.randint(1, 11, (3, 3))

producto = np.dot(matriz1, matriz2)

print("Ejercicio 7:")
print("Matriz 1:")
print(matriz1)
print("Matriz 2:")
print(matriz2)
print("Producto de matrices:")
print(producto)
```

## Ejercicio 8
**Crear un array de 15 valores aleatorios entre 1 y 100 e imprimir su máximo, mínimo y promedio.**

```python
array_random = np.random.randint(1, 101, 15)

print("Ejercicio 8:")
print("Array:", array_random)
print("Máximo:", np.max(array_random))
print("Mínimo:", np.min(array_random))
print("Promedio:", np.mean(array_random))
```

## Ejercicio 9
**Generar una matriz 5x5 con valores del 1 al 25 y extraer su diagonal.**

```python
matriz_1_25 = np.arange(1, 26).reshape(5, 5)
diagonal = np.diag(matriz_1_25)

print("Ejercicio 9:")
print("Matriz:")
print(matriz_1_25)
print("Diagonal:")
print(diagonal)
```

## Ejercicio 10
**Dada una matriz 5x5 de enteros aleatorios entre 1 y 20, calcular la suma de cada fila y cada columna.**


```python
matriz_random = np.random.randint(1, 21, (5, 5))

suma_filas = np.sum(matriz_random, axis=1)
suma_columnas = np.sum(matriz_random, axis=0)

print("Ejercicio 10:")
print("Matriz:")
print(matriz_random)
print("Suma de cada fila:")
print(suma_filas)
print("Suma de cada columna:")
print(suma_columnas)
```
