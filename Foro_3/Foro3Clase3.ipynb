1.    Cargar el conjunto de datos en una estructura adecuada para su manipulación.

```python
import pandas as pd

ruta = r"C:\Users\Unicom-7344\Desktop\Emiliano_Palacios\Uni 2\Aprendizaje Automatico\Automobile.csv"

df = pd.read_csv(ruta)

df.head()
```

2.   Crear un gráfico de dispersión para visualizar las relaciones entre los pares de características numéricas `mpg` y `weight`.

```python
import matplotlib.pyplot as plt
```

```python
plt.figure()

plt.scatter(df['weight'], df['mpg'], label='Datos', marker='o')

plt.xlabel('Peso del automóvil')
plt.ylabel('Consumo (mpg)')
plt.title('Relación entre peso y consumo')

plt.legend(loc='upper right')
plt.grid()

plt.xlim(df['weight'].min(), df['weight'].max())
plt.ylim(df['mpg'].min(), df['mpg'].max())

plt.show()
```

3.   Generar un histograma del consumo de combustible (`mpg`) para analizar su distribución.

```python
plt.figure()

plt.hist(df['mpg'], bins=15, label='Distribución mpg',
         edgecolor='black', linewidth=1)

plt.xlabel('Millas por galón (mpg)')
plt.ylabel('Frecuencia')
plt.title('Distribución del consumo')

plt.legend(loc='upper right')
plt.grid(False)

plt.show()
```

4.   Graficar un diagrama de caja (boxplot) de las características `cylinders` y `origin` para identificar posibles valores atípicos.

```python
plt.figure(figsize=(8,5))

df.boxplot(column="cylinders", by="origin", grid=False)

plt.xlabel("Origen del automóvil")
plt.ylabel("Cantidad de cilindros")
plt.title("Distribución de cilindros según el origen")

plt.xticks([1,2,3], ['USA', 'Europa', 'Japón'])

plt.suptitle("") 

plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```

5.   Realizar un gráfico de barras que muestre la distribución de automóviles según su año de modelo (`model_year`) y origen (`origin`).

```python
tabla = df.groupby(['model_year', 'origin']).size().unstack()

plt.figure()

tabla.plot(kind='bar')

plt.xlabel('Año del modelo')
plt.ylabel('Cantidad de autos')
plt.title('Distribución por año y origen')

plt.legend(title='Origen', loc='upper right')
plt.grid()

plt.show()
```
